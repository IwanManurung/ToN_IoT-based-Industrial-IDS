### Read IoT Dataset

```python
df = pd.read_csv('Processed_IoT_dataset/IoT_Modbus.csv', low_memory=False)
```

```python
df.dropna(subset='date', inplace=True)
df.dropna(subset='time', inplace=True)
df['time'] = df['time'].str.strip()
df['date'] = df['date'].str.strip()
df['date'] = df['date'].str.replace('Apr-19', 'April 2019')
df['date'] = df['date'].str.replace('Mar-19', 'March 2019')
df['date'] = df['date'].str.replace('-', ' ')
df['dt'] = df[['date', 'time']].agg(' '.join, axis=1)
df['dt'] = df['dt'].str.strip()
df['dt'] = df['dt'].str.replace('  ', ' ')
df['dt'] = pd.to_datetime(df['dt'], format='%d %B %Y %H:%M:%S')
df['timestamp'] = df['dt'].astype(int)/10**9
df.sort_values('timestamp', inplace=True)
df.reset_index(drop=True, inplace=True)
df.dropna(inplace=True)
```
