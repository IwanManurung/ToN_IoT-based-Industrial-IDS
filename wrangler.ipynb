## Drift Detectors

```python
import numpy as np
import pandas as pd
# import matplotlib.pyplot as plt
# import time
from base import params
# from src import iids_util
```

```python
df = pd.read_csv('datastream/ToN_IoT_Edge_dataset/Edge_Fusion_IoT_Fridge.csv')
```

```python
for col in df.drop(columns=params.non_feature).columns:
    maks = df[col].max()
    if maks >= 1_000:
        df[col] = np.divide(df[col], 1_000_000)
    df[col] = df[col].astype(np.float16)
    msg = f"{col}. Min: {df[col].min()}. Max: {df[col].max()}. Unique: {len(df[col].unique())}"
    print(msg)
```
